```python
# Filling missing value with NAN value
data_copy = data.copy (deep = True)
data_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = data_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
data_copy
```

```python
# checking NAN values in dataset
data_copy.isnull().sum()
```

```python
# visualizing data before filling null values
import matplotlib.pyplot as plt
d = data_copy.hist(figsize = (12,12))
```

```python
# imputing null values with the mean value of each column
data_copy['Glucose'].fillna(data_copy['Glucose'].mean(), inplace = True)
data_copy['BloodPressure'].fillna(data_copy['BloodPressure'].mean(), inplace = True)
data_copy['SkinThickness'].fillna(data_copy['SkinThickness'].mean(), inplace = True)
data_copy['Insulin'].fillna(data_copy['Insulin'].mean(), inplace = True)
data_copy['BMI'].fillna(data_copy['BMI'].mean(), inplace = True)
```

```python
# Ploting distribution after removing NaN value
df = data_copy.hist(figsize= (12,12))
```

Building Model

```python
# splitting the data
x = data_copy.drop('Outcome', axis = 1)
y = data_copy['Outcome']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.33,
                                                    random_state=7)
```

# **Random Forest Model**

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200)
rf.fit(X_train, y_train)
```

Checking accuracy of the model in the traing set.

```python
rf_train = rf.predict(X_train)
from sklearn import metrics

print("Accuracy_Score =", format(metrics.accuracy_score(y_train, rf_train)))
```

Since the accuracy score is 1, the training dataset is overfitted.

```python
# accuracy of the test data
predictions = rf.predict(X_test)
print("Accuracy_Score =", format(metrics.accuracy_score(y_test, predictions)))
```

```python
print (confusion_matrix(y_test, predictions))
print( classification_report(y_test, predictions))
```
